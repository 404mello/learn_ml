```python
"""
There are two plots that show the interactions between multiple variables in your dataset:
    Correlation Matrix plot.
    Scatter Plot Matrix.
"""
print("Definition of Multivariate Plots")
```

```python
"""
CORRELATION MATRIX PLOT
Correlation gives and indication of how related the changes are between two variables.
If two variables change in the same direction they are positively correlated.
If they can change in opposite directions together (inversely proportional), then they are negatively correlated.
You vcan calculate the correaltion between each pair of attributes.
This is called a correlation matrix.
"""
print("Definition of a corelation matrix")
```

```python
"""
Correlation matrix plot
"""
#import matplotlib.pyplot as plt
from pandas import read_csv
import numpy as np
filename = "winequality-red.csv"

data = read_csv(filename )
names = list(data.columns)
#Method to calculate correlation
correlations = data.corr()
#Plot correlation matrix

fig = plt.figure()
#Subplot is used for axes- multiple arguements
ax = fig.add_subplot(111) #Single three digit arguement 1-x axis, 2-y axis, 3-position.
#Syntax to display the matrix 
cax = ax.matshow(correlations, vmin=-1, vmax=1) #Max and min cor values

fig.colorbar(cax) #Assign colour to matrix
#Partitioning labels according to 1 each.
ticks = np.arange(0,12,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

ax.set_xticklabels(names)
ax.set_yticklabels(names)
plt.show()
```

```python
"""
SCATTER PLOT MATRIX:
A scatter plot shows the relationship between two variables as dots in two dimensions, one axis for each attribute. 
You can create a scatter plot for each pair of attributes in your data. Drawing all these scatter plots together is called a scatter plot matrix.
"""
from pandas.plotting import scatter_matrix
filename = "winequality-red.csv"
data = read_csv(filename )
names = list(data.columns)
scatter_matrix(data)
plt.show()

print("Scatter plot")
```
